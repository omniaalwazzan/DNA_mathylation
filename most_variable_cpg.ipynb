###  Method 1: Coefficient of Variation (CV):

```python
#
cv = (df_dna_850k.std(axis=1) / df_dna_850k.mean(axis=1)) * 100  # CV as a percentage
number_of_probes=8000
most_variable_probes_cv = cv.nlargest(n=number_of_probes)


```

###  Method 2: Interquartile Range (IQR): Calculate the interquartile range (IQR) for each probe. IQR is the range between the 25th and 75th percentiles.


```python
iqr = df.quantile(0.75, axis=1) - df.quantile(0.25, axis=1)
most_variable_probes_iqr = iqr.nlargest(n=8000)
```

### save selected cpg obtained from IQR to a csv

```python
# M-Val_path = "/data/DERI-MMH/DNA_meth/selected_Mval_feat/iqr_cpg_mVal.csv"
iqr_beta_path =  "/data/DERI-MMH/DNA_meth/selected_beta_feat/iqr_cpg_feat.csv"

series2 = most_variable_probes_iqr.to_frame().reset_index(drop=False).to_csv(iqr_beta_path)
```

###  Method 3: Median Absolute Deviation (MAD):


```python
mad = df.mad(axis=1)
most_variable_probes_mad = mad.nlargest(n=8000)
```

### save selected cpg obtained from MAD to a csv

```python
# M-Val_path = "/data/DERI-MMH/DNA_meth/selected_Mval_feat/iqr_cpg_mVal.csv"
mad_beta_path =  "/data/DERI-MMH/DNA_meth/selected_beta_feat/mad_cpg_feat.csv"
series3 = most_variable_probes_mad.to_frame().reset_index(drop=False).to_csv(mad_beta_path)
```

### Method 4: Standard Deviation or Variance


```python
variability = df.var(axis=1)  # Calculate variance across rows (probes)
most_variable_probes_vr = variability.nlargest(n=number_of_probes)
```

### save selected cpg obtained from Var to a csv


```python
# M-Val_path = "/data/DERI-MMH/DNA_meth/selected_Mval_feat/iqr_cpg_mVal.csv"
var_beta_path =  "/data/DERI-MMH/DNA_meth/selected_beta_feat/var_cpg_feat.csv"
series4 = most_variable_probes_vr.to_frame().reset_index(drop=False).to_csv(var_beta_path)
```

```python
# Saving all selected feature to one csv
series1 = most_variable_probes_cv.to_frame().reset_index(drop=False)
series2 = most_variable_probes_iqr.to_frame().reset_index(drop=False)
series3 = most_variable_probes_mad.to_frame().reset_index(drop=False)
series4 = most_variable_probes_vr.to_frame().reset_index(drop=False)

concatenated_df = pd.concat([series1, series2,series3,series4], axis=1, ignore_index=True)
concatenated_df = concatenated_df.rename(columns={0: 'CpG_cv', 1: 'score',2:'CpG_iqr',3:'score',4:'CpG_mad',5:'socre',6:'CpG_vr',7:'score'})
concatenated_df
```
